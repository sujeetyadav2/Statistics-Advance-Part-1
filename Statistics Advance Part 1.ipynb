## Theoretical Questions ##


**Q1. What is a random variable in probability theory?**

 Ans- A random variable is a function that maps outcomes of a random experiment to numerical values. It provides a mathematical way to quantify uncertainty.
 
Example: When rolling a die, the random variable X can take values {1, 2, 3, 4, 5, 6}, each with probability 1/6.

**Q2. What are the types of random variables?**

 Ans- There are two main types:

  1. Discrete:- Takes countable, finite values

   ex. Number of heads in 10 flips

  3. Continuous:- Takes any value in a range

   ex. Height, weight, temperature

 A third type, Mixed, combines both discrete and continuous behavior (e.g., insurance payouts).



**Q3. What is the difference between discrete and continuous distributions?**

Discrete Distribution:- 

Takes countable integer values, uses PMF, and P(X = x) can be greater than 0. Total probability is calculated as Σ P(x) = 1

Example: Binomial, Poisson.
    
Continuous Distribution :-

Takes any real number in a range, uses PDF, and P(X = x) is always exactly 0. Total probability is calculated as ∫ f(x)dx = 1. 
    
Example: Normal, Uniform.

**Q4. What are probability distribution functions (PDF)?**

 Ans- A Probability Distribution Function describes how probabilities are spread over the values of a random variable.
For discrete variables it is called PMF (Probability Mass Function) and gives P(X = x) directly.
    
For continuous variables it is called PDF (Probability Density Function) and gives density f(x), where probability over a range = ∫f(x)dx.
    
Properties: f(x) ≥ 0 for all x, and total area under the curve = 1.

**Q5. How do cumulative distribution functions (CDF) differ from PDFs?**

 Ans- PDF/PMF gives probability at a specific value. Its output can be any non-negative number and it is not monotonic.
     
CDF gives F(x) = P(X ≤ x), meaning probability up to point x. Its output always lies between 0 and 1 and it is always non-decreasing.
    
The key relationship is: CDF is the integral of PDF, and PDF is the derivative of CDF.
    
Key point: PDF tells "probability at a point", CDF tells "probability up to a point."

**Q6. What is a discrete uniform distribution?**

 Ans- A distribution where all outcomes are equally likely over a finite set of values.
     
PMF: P(X = x) = 1/n, for x ∈ {x₁, x₂, ..., xₙ}

Mean = (a + b) / 2

Variance = (n² − 1) / 12

Example: Rolling a fair die — each of {1, 2, 3, 4, 5, 6} has probability 1/6.

**Q7. What are the key properties of a Bernoulli distribution?**

 Ans- A Bernoulli distribution models a single trial with only two outcomes: success (1) or failure (0).

PMF: P(X=1) = p and P(X=0) = 1−p

Mean = p

Variance = p(1−p)

Support = {0, 1}

It is the building block of the Binomial distribution.
    
Example: Single coin flip with p = 0.5 where Heads = success.

**Q8. What is the binomial distribution, and how is it used in probability?**

 Ans- The Binomial distribution models the number of successes in n independent Bernoulli trials.

PMF: P(X = k) = C(n,k) × pᵏ × (1−p)ⁿ⁻ᵏ

Mean = np

Variance = np(1−p)

Parameters: n = number of trials, p = probability of success

Applications:- Quality control, medical trials, election polling, spam detection.

**Q9. What is the Poisson distribution and where is it applied?**

 Ans- The Poisson distribution models the number of events occurring in a fixed interval of time or space, given a known average rate λ.
    
PMF: P(X = k) = (e⁻λ × λᵏ) / k!

Mean = Variance = λ

Applications:

1. Number of calls arriving at a call center per hour
2. Number of emails received per day
3. Number of road accidents per week



**Q10. What is a continuous uniform distribution?**

 Ans- A distribution where all values in range [a, b] are equally likely.
     
PDF: f(x) = 1/(b−a), for a ≤ x ≤ b
Mean = (a + b) / 2
Variance = (b−a)² / 12
CDF: F(x) = (x−a)/(b−a)

Example: Random number generation between 0 and 1.

**Q11. What are the characteristics of a normal distribution?**

 Ans- The normal (Gaussian) distribution is the most widely used continuous distribution.
 
It has a symmetric, bell-shaped curve where Mean = Median = Mode. It is defined by two parameters: μ (mean) and σ (standard deviation).

PDF: f(x) = (1/σ√2π) × e^[−(x−μ)²/2σ²]

Empirical Rule:

1. 68% of data falls within 1σ of the mean
2. 95% of data falls within 2σ of the mean
3. 99.7% of data falls within 3σ of the mean

The tails extend to ±∞ but never touch zero.

**Q12. What is the standard normal distribution, and why is it important?**

 Ans- The standard normal distribution is a special case of normal distribution with μ = 0 and σ = 1, denoted as Z ~ N(0, 1).

Why it is important:

Allows comparison across different distributions on a common scale

1. Enables use of Z-tables for probability calculations
2. Foundation for hypothesis testing and confidence intervals
3. Any normal distribution can be standardized using: Z = (X − μ) / σ

**Q13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?**

 Ans- The Central Limit Theorem states that the sampling distribution of the sample mean approaches a normal distribution as sample size n increases (generally n ≥ 30), regardless of the population's original shape.

Why it is critical:

1. Justifies use of Z-tests and t-tests on real data
2. Makes normal distribution applicable to almost all real-world problems
3. Enables valid inference about population parameters from sample data
4. It is the theoretical foundation of most inferential statistics

**Q14. How does the Central Limit Theorem relate to the normal distribution?**

 Ans- CLT is the theoretical reason why normal distribution appears everywhere in statistics.
 
For large n: X̄ ~ N(μ, σ²/n) approximately, regardless of the original population shape.

Even if the population is skewed, exponential, or uniform, the sample means will form a normal distribution. The larger the sample size, the better the normal approximation becomes.

In short, CLT essentially "creates" normality from any distribution through repeated sampling.

**Q15. What is the application of Z statistics in hypothesis testing?**

 Ans- Z-statistics are used when population standard deviation σ is known and sample size is large (n ≥ 30).

Steps in Z-test:

1. State H₀ (null hypothesis) and H₁ (alternate hypothesis)
2. Calculate Z = (X̄ − μ₀) / (σ/√n)
3. Compare Z to critical value (e.g., ±1.96 for 95% confidence level)
4. If |Z| > critical value then Reject H₀, otherwise Fail to Reject H₀

**Q16. How do you calculate a Z-score, and what does it represent?**

 Ans- Formula: Z = (X − μ) / σ
 
where X is the observed value, μ is the population mean, and σ is the population standard deviation.

Z tells us how many standard deviations the value X is away from the mean.

Interpretation:

1. Z = 0 means the value equals the mean
2. Z = 2 means the value is 2 standard deviations above the mean
3. Z = −1.5 means the value is 1.5 standard deviations below the mean

**Q17. What are point estimates and interval estimates in statistics?**

 Ans- Point Estimate is a single value used to estimate a population parameter. For example, sample mean X̄ is used to estimate population mean μ. It is simple but does not express uncertainty.
 
Interval Estimate is a range of values likely to contain the true population parameter. For example, X̄ ± 1.96×(σ/√n) gives a 95% confidence interval. It is more informative and reliable than a point estimate. 

**Q18. What is the significance of confidence intervals in statistical analysis?**

 Ans- A confidence interval (CI) gives a range within which the true population parameter lies with a specified probability called the confidence level.

Formula for 95% CI: X̄ ± 1.96 × (σ/√n)

Significance:

1. Quantifies the uncertainty in our estimation
2. A 95% CI means if we repeated the experiment 100 times, approximately 95 of those intervals would contain the true parameter
3. Wider CI means more uncertainty, Narrower CI means more precision
4. Essential in medical research, scientific studies, and business decisions

**Q19. What is the relationship between a Z-score and a confidence interval?**

 Ans- The Z-score defines the critical value that determines how wide the confidence interval is.
 
1. For 90% confidence level the critical Z-score is 1.645.
2. For 95% confidence level the critical Z-score is 1.960.
3. For 99% confidence level the critical Z-score is 2.576.

CI formula: X̄ ± Z* × (σ/√n)

Higher confidence level means larger Z* which means wider interval. The Z-score is the bridge between a confidence level and its corresponding interval boundaries. 

**Q20. How are Z-scores used to compare different distributions?**

 Ans- Z-scores standardize data to a common scale, making meaningful cross-distribution comparisons possible.
 
Example:

1.Math: Score = 80, μ = 70, σ = 5 → Z = (80−70)/5 = 2.0

2. Science: Score = 75, μ = 60, σ = 6 → Z = (75−60)/6 = 2.5

Even though the raw Science score is lower, the Z-score shows the student performed relatively better in Science.

This is the power of Z-score standardization — it removes the effect of different scales and units.

**Q21. What are the assumptions for applying the Central Limit Theorem?**

 Ans- Independence: Each sample observation must be independent of others.
Random Sampling:- Data must be randomly drawn from the population.

Sufficient Sample Size:- Generally n ≥ 30 is required. For highly skewed distributions a larger n may be needed.

Finite Variance:- The population must have a finite mean and finite variance.

Identical Distribution:- All samples must be drawn from the same population distribution.                                

**Q22. What is the concept of expected value in a probability distribution?**

 Ans- The expected value E[X] is the long-run average value of a random variable. It is the weighted average of all possible outcomes, weighted by their probabilities.
 
For Discrete: E[X] = Σ [x × P(X = x)]

For Continuous: E[X] = ∫ x × f(x) dx

Example: 
For a fair die:-
E[X] = 1×(1/6) + 2×(1/6) + 3×(1/6) + 4×(1/6) + 5×(1/6) + 6×(1/6) = 3.5

This means if we roll a die infinitely many times, the average result will approach 3.5.


**Q23. How does a probability distribution relate to the expected outcome of a random variable?**

 Ans- The probability distribution fully describes the behavior of a random variable, and the expected value is derived directly from it.
 
The distribution assigns probabilities to all possible outcomes. The expected value is the probability-weighted center of that distribution. Outcomes with higher probability contribute more to E[X].

E[X] represents the theoretical mean we would observe over infinite repetitions of the experiment.

Variance is also derived from the distribution: Var(X) = E[X²] − (E[X])²
In short, the distribution is the complete picture and expected value is one key summary of that picture.

```python
#Q3. Write a Python function to calculate the PDF of a Bernoulli distribution.

import matplotlib.pyplot as plt

def bernoulli_pmf(p, x):
    if x == 1:
        return p
    elif x == 0:
        return 1 - p
    else:
        raise ValueError("x must be 0 or 1")

p = 0.6
print(f"P(X=1) = {bernoulli_pmf(p, 1)}")
print(f"P(X=0) = {bernoulli_pmf(p, 0)}")

outcomes = [0, 1]
probs = [bernoulli_pmf(p, x) for x in outcomes]

plt.bar(outcomes, probs, color=['tomato', 'steelblue'],
        width=0.4, edgecolor='black')
plt.xticks([0, 1], ['Failure (0)', 'Success (1)'])
plt.ylabel('Probability')
plt.title(f'Bernoulli Distribution PMF (p = {p})')
plt.tight_layout()
plt.show()
```

```python
#Q9. Write a Python function to calculate Z-scores from a dataset and plot them.

import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    return (data - mean) / std

data = np.array([23, 45, 67, 12, 89, 56, 34, 78, 90, 43])
z_scores = calculate_z_scores(data)

print("Original Data:", data)
print("Z-Scores:     ", np.round(z_scores, 2))

colors = ['tomato' if z < 0 else 'steelblue' for z in z_scores]
plt.figure(figsize=(10, 5))
plt.bar(range(len(data)), z_scores, color=colors, edgecolor='black')
plt.axhline(0, color='black', linewidth=1)
plt.xlabel('Data Point Index')
plt.ylabel('Z-Score')
plt.title('Z-Scores of Dataset')
plt.tight_layout()
plt.show()
```

```python
#Q14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.


import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def z_score(x, mu, sigma):
    return (x - mu) / sigma

x_val, mu, sigma = 75, 60, 10
z = z_score(x_val, mu, sigma)
print(f"Z-score for X={x_val}: Z = {z}")
print(f"P(Z <= {z}) = {norm.cdf(z):.4f}")

x = np.linspace(-4, 4, 300)
plt.figure(figsize=(8, 5))
plt.plot(x, norm.pdf(x), 'b-', lw=2, label='Standard Normal PDF')
plt.axvline(z, color='red', linestyle='--', label=f'Z = {z}')
plt.fill_between(x, norm.pdf(x), where=(x <= z),
                 alpha=0.3, color='red', label='P(Z <= z)')
plt.title(f'Z-score = {z} on Standard Normal Distribution')
plt.xlabel('Z')
plt.ylabel('Density')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Q15. Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
from scipy import stats

sample = np.random.normal(105, 15, 50)
mu_0 = 100
sigma = 15
n = len(sample)
x_bar = np.mean(sample)

z_stat = (x_bar - mu_0) / (sigma / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print(f"Sample Mean   : {x_bar:.4f}")
print(f"Z-statistic   : {z_stat:.4f}")
print(f"P-value       : {p_value:.4f}")
print("Decision      :", "Reject H0" if p_value < 0.05 else "Fail to Reject H0")
```

```python
#Q22. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.

import numpy as np

data = np.random.normal(100, 20, 300)

range_ = np.max(data) - np.min(data)
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

print(f"Min    : {np.min(data):.2f}")
print(f"Max    : {np.max(data):.2f}")
print(f"Range  : {range_:.2f}")
print(f"Q1     : {q1:.2f}")
print(f"Q3     : {q3:.2f}")
print(f"IQR    : {iqr:.2f}")
```

```python
#Q23. Implement Z-score normalization on a dataset and visualize its transformation.

import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(50, 10, 300)
z_normalized = (data - np.mean(data)) / np.std(data)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].hist(data, bins=30, color='orange', edgecolor='black')
axes[0].set_title(f'Original Data (μ={np.mean(data):.1f}, σ={np.std(data):.1f})')

axes[1].hist(z_normalized, bins=30, color='green', edgecolor='black')
axes[1].set_title('Z-Normalized Data (μ=0, σ=1)')

for ax in axes:
    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
#Q24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

def analyze_distribution(data):
    sk = skew(data)
    kurt = kurtosis(data)
    print(f"Mean     : {np.mean(data):.4f}")
    print(f"Std Dev  : {np.std(data):.4f}")
    print(f"Skewness : {sk:.4f}")
    print(f"Kurtosis : {kurt:.4f}")

    if abs(sk) < 0.5:
        print("Distribution is approximately symmetric")
    else:
        print("Distribution is skewed")

    if abs(kurt) < 0.5:
        print("Distribution has normal tails (mesokurtic)")
    elif kurt > 0.5:
        print("Distribution has heavy tails (leptokurtic)")
    else:
        print("Distribution has light tails (platykurtic)")

data = np.random.normal(0, 1, 1000)
analyze_distribution(data)

plt.hist(data, bins=40, color='skyblue', edgecolor='black', density=True)
plt.title('Normal Distribution — Skewness and Kurtosis Analysis')
plt.xlabel('Value')
plt.ylabel('Density')
plt.tight_layout()
plt.show()
```
